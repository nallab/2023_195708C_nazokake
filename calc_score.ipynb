```python
def score(tokens_tensor):
  loss=model(tokens_tensor, labels=tokens_tensor)[0]
  return np.exp(loss.cpu().detach().numpy())
def calc_score(text):
  tokens_tensor = tokenizer.encode(text,add_special_tokens=False, return_tensors="pt")
  return score(tokens_tensor)
```

```python
def Nazokake_For_AI(nazokake):
  nazokake = nazokake.replace("その心は、","").replace("かけまして、","").replace("とときます。","には、").replace("どちらも、","どちらも")
  nazokake_for_ai = ""
  #nazokake_for_ai = nazokake
  in_kakko = True
  for i in nazokake:
    if(i=="("):
      in_kakko = False
    elif(i==")"):
      in_kakko = True
    elif(in_kakko):
      nazokake_for_ai += i
  return nazokake_for_ai
```

```python
predict_nazokake = original
score_list = []
for text in predict_nazokake:
  nazokake_for_AI = Nazokake_For_AI(text)
  
  text_score = calc_score(nazokake_for_AI)
  print(nazokake_for_AI)
  print(str(text_score))
  score_list.append(text_score)
```

```python
def max_index(_list):
  save_list = []
  index_list = []
  for i in _list:
    save = 0
    index = 0
    for k,l in enumerate(i):
      if k==0:
        save = l
        index = k
      elif l < save:
        save = l
        index = k
    save_list.append(save)
    index_list.append(index)
  return save_list,index_list
```
